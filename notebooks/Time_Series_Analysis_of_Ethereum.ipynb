**ARIMA(Autoregressive Integrated Moving Average)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load Etherium data
eth = pd.read_excel("/content/ethereum_unnormalized.xlsx")

# Set Date as index and ensure datetime format
eth['Date'] = pd.to_datetime(eth['Date'])
eth.set_index('Date', inplace=True)

# Extract 'Close' prices and set frequency
eth_close = eth['ETH-USD_Close'].asfreq('D')
eth_close = eth_close.fillna(method='ffill')

# Auto ARIMA to find best parameters
from pmdarima import auto_arima
auto_model = auto_arima(eth_close, seasonal=False, trace=True, error_action='ignore', suppress_warnings=True)

# Fit ARIMA model
arima_model = ARIMA(eth_close, order=auto_model.order)
arima_result = arima_model.fit()

# Forecast next 5 days
forecast_arima = arima_result.forecast(steps=5)

# Plotting ARIMA forecast
plt.figure(figsize=(14, 6))
plt.plot(eth_close[-100:], label='Actual (Last 100 Days)', linewidth=2)
plt.plot(forecast_arima, label='ARIMA Forecast (5 days)', color='blue')
plt.title('Ethereum Price Forecast: ARIMA Model')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**SARIMA(Seasonal Autoregressive Integrated Moving Average)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load Etherium data
eth = pd.read_excel("/content/ethereum_unnormalized.xlsx")

# Set Date as index and ensure datetime format
eth['Date'] = pd.to_datetime(eth['Date'])
eth.set_index('Date', inplace=True)

# Extract 'Close' prices and set frequency
eth_close = eth['ETH-USD_Close'].asfreq('D')
eth_close = eth_close.fillna(method='ffill')

# Define SARIMA order parameters manually
# (p, d, q) for ARIMA part
order = (1, 1, 1)

# (P, D, Q, s) for seasonal part
seasonal_order = (1, 1, 1, 7)  # weekly seasonality (s=7)

# Fit SARIMA model
sarima_model = SARIMAX(
    eth_close,
    order=order,
    seasonal_order=seasonal_order,
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarima_result = sarima_model.fit(disp=False)

# Forecast next 30 days
forecast_sarima = sarima_result.get_forecast(steps=30)
forecast_index = pd.date_range(start=eth_close.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')
forecast_series = pd.Series(forecast_sarima.predicted_mean.values, index=forecast_index)

# Plotting SARIMA forecast
plt.figure(figsize=(14, 6))
plt.plot(eth_close[-100:], label='Actual (Last 100 Days)', linewidth=2)
plt.plot(forecast_series, label='SARIMA Forecast (30 days)', color='red')
plt.title('Ethereum Price Forecast: SARIMA Model')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Exponential Smoothing (Holt-Winters Method)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Load Ethereum data
eth = pd.read_excel("/content/ethereum_unnormalized.xlsx")

# Set Date as index and ensure datetime format
eth['Date'] = pd.to_datetime(eth['Date'])
eth.set_index('Date', inplace=True)

# Extract 'Close' prices and set frequency
eth_close = eth['ETH-USD_Close'].asfreq('D')
eth_close = eth_close.fillna(method='ffill')

# Fit Holt-Winters Model
hw_model = ExponentialSmoothing(eth_close, trend='add', seasonal=None, damped_trend=True)
hw_result = hw_model.fit()

# Forecast next 3 days
forecast_hw = hw_result.forecast(3)

# Plotting Holt-Winters forecast
plt.figure(figsize=(14, 6))
plt.plot(eth_close[-100:], label='Actual (Last 100 Days)', linewidth=2)
plt.plot(forecast_hw, label='Holt-Winters Forecast (3 days)', color='green')
plt.title('Ethereum Price Forecast: Holt-Winters Model')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Random Forest Regression**

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import joblib
import os

# Create models folder if it doesn't exist
os.makedirs('models', exist_ok=True)

# Load dataset
df = pd.read_excel('/content/ethereum_normalized.xlsx', parse_dates=['Date'])
df.sort_values('Date', inplace=True)

# Create target variable: next day's Close price
df['Target'] = df['ETH-USD_Close'].shift(-1)

# Drop last row with NaN target
df.dropna(inplace=True)

# Define features and target
features = ['ETH-USD_Open', 'ETH-USD_High', 'ETH-USD_Low', 'ETH-USD_Close',
            'ETH-USD_Volume', 'Volatility', 'MA_7', 'MA_30', 'Daily_Return', 'Sentiment_Score']

X = df[features]
y = df['Target']

# Time series train-test split (89% train)
split_index = int(len(df) * 0.89)
X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]

print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# Train Random Forest Regression
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict and evaluate
rf_preds = rf_model.predict(X_test)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_preds))

print(f"Random Forest RMSE: {rf_rmse:.4f}")

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(y_test.values, label='Actual Close Price', color='blue')
plt.plot(rf_preds, label='Predicted Close Price', color='red')
plt.title('Random Forest: Actual vs Predicted Close Price')
plt.xlabel('Time (Days)')
plt.ylabel('ETH-USD_Close Price')
plt.legend()
plt.show()

import os
os.makedirs('models', exist_ok=True)

# Save ETH model
joblib.dump(rf_model, 'models/rf_eth_model.pkl')
```

**LSTM (Long Short-Term Memory) Networks**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Load your dataset
df = pd.read_excel('/content/ethereum_normalized.xlsx', parse_dates=['Date'])
df.sort_values('Date', inplace=True)

# Create target: next day Close price
df['Target'] = df['ETH-USD_Close'].shift(-1)
df.dropna(inplace=True)

# Features for LSTM
features = ['ETH-USD_Open', 'ETH-USD_High', 'ETH-USD_Low', 'ETH-USD_Close',
            'ETH-USD_Volume', 'Volatility', 'MA_7', 'MA_30', 'Daily_Return', 'Sentiment_Score']

X = df[features]
y = df['Target']

# Scale data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))

# Create time sequences for LSTM
def create_sequences(X, y, time_steps=10):
    X_seq, y_seq = [], []
    for i in range(len(X) - time_steps):
        X_seq.append(X[i:i+time_steps])
        y_seq.append(y[i+time_steps])
    return np.array(X_seq), np.array(y_seq)

time_steps = 10
X_seq, y_seq = create_sequences(X_scaled, y_scaled, time_steps)

# Train-test split (80%)
train_size = int(len(X_seq) * 0.8)
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]

print(f"LSTM Train shape: {X_train.shape}, Test shape: {X_test.shape}")

# Build LSTM Model
model = Sequential([
    LSTM(64, activation='relu', return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.summary()

# Train model
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1, verbose=1)

# Predict
y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_test_actual = scaler_y.inverse_transform(y_test)

# Evaluate RMSE
lstm_rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred))
print(f"LSTM RMSE: {lstm_rmse:.4f}")

# Plot Actual vs Predicted
plt.figure(figsize=(12,6))
plt.plot(y_test_actual, label='Actual Close Price', color='blue')
plt.plot(y_pred, label='Predicted Close Price', color='red')
plt.title('LSTM: Actual vs Predicted Close Price')
plt.xlabel('Time (Days)')
plt.ylabel('ETH-USD_Close Price')
plt.legend()
plt.show()
```

**Recurrent Neural Networks (RNNs)**

```python
# Install TensorFlow
!pip install tensorflow

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Dropout

# Load and sort data
df = pd.read_excel('/content/ethereum_normalized.xlsx', parse_dates=['Date'])
df.sort_values('Date', inplace=True)

# Target: next day's close
df['Target'] = df['ETH-USD_Close'].shift(-1)
df.dropna(inplace=True)

# Feature columns
features = ['ETH-USD_Open', 'ETH-USD_High', 'ETH-USD_Low', 'ETH-USD_Close',
            'ETH-USD_Volume', 'Volatility', 'MA_7', 'MA_30', 'Daily_Return', 'Sentiment_Score']

X = df[features]
y = df['Target']

# Normalize
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))

# Create sequences
def create_sequences(X, y, time_steps=10):
    X_seq, y_seq = [], []
    for i in range(len(X) - time_steps):
        X_seq.append(X[i:i+time_steps])
        y_seq.append(y[i+time_steps])
    return np.array(X_seq), np.array(y_seq)

time_steps = 10
X_seq, y_seq = create_sequences(X_scaled, y_scaled, time_steps)

# Train-test split
split = int(len(X_seq) * 0.5)
X_train, X_test = X_seq[:split], X_seq[split:]
y_train, y_test = y_seq[:split], y_seq[split:]

print(f"RNN shape - Train: {X_train.shape}, Test: {X_test.shape}")

# Build RNN model
model = Sequential([
    SimpleRNN(64, activation='tanh', input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')
model.summary()

# Train model
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1, verbose=1)

# Predict
y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_test_actual = scaler_y.inverse_transform(y_test)

# Plot Actual vs Predicted
test_years = df['Date'].iloc[-len(y_test_actual):].dt.year.values

plt.figure(figsize=(12,6))
plt.plot(y_test_actual, label='Actual Close Price', color='blue')
plt.plot(y_pred, label='Predicted Close Price', color='red')
plt.title('RNN: Actual vs Predicted Close Price')
plt.xlabel('Time(Days)')
plt.ylabel('ETH-USD_Close Price')
plt.legend()
plt.show()

# Evaluate RMSE
rnn_rmse = np.sqrt(mean_squared_error(y_test_actual, y_pred))
print(f'RNN RMSE: {rnn_rmse:.4f}')
```
